### What is a String
Definition:
A string is a sequence of characters used to store text data.
In Python, strings are written inside single quotes ' ' or double quotes " ".    

üß™ Example

```python
name = "Python"
print(name)
```

üîë Important Points for Students

Strings are immutable (cannot be changed)

Indexing starts from 0

Strings support many built-in methods

### String Indexing
üìò Definition

Indexing means accessing individual characters in a string.

```python
word = "Python"
print(word[0])   # P
print(word[3])   # h
```

### String Slicing
üìò Definition

Slicing is used to get a part of a string.

```python
word = "Python"
print(word[0:4])   # Pyth
print(word[2:])    # thon
print(word[:3])    # Pyt
```

### String Operations
1. ‚ûï String Concatenation

Definition:
Joining two or more strings

```python
a = "Hello"
b = "World"
print(a + " " + b)
```

2.  String Repetition

Definition:
Repeating a string multiple times.

```python
print("Hi " * 3)
```

3. üîç Membership Operators

Definition:
Check if a character or word exists in a string.

```python
text = "Python"
print("P" in text)
print("z" not in text)
```

### Common String Methods
üî† upper()

Definition: Converts string to uppercase.

```python
text = "python"
print(text.upper())
```

2. üî° lower()

Definition: Converts string to lowercase.

```python
text = "PYTHON"
print(text.lower())
```

3. üßπ strip()

Definition: Removes spaces from start and end.

```python
text = "  hello  "
print(text.strip())
```

4. üîÑ replace()

Definition: Replaces one word/character with another.

```python
text = "I like Java"
print(text.replace("Java", "Python"))
```

5. ‚úÇ split()

Definition: Splits string into a list.

```python
text = "Python is easy"
print(text.split())
```

6. String Formatting\
üìò Definition

Used to insert variables into strings.

```python
name = "Ali"
age = 20
print(f"My name is {name} and I am {age} years old")
```

7. Reverse a String
üìò Definition

Reverse means reading string backward.

```python
text = "Python"
print(text[::-1])
```
