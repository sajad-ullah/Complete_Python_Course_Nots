### 1. What is a Function in Python

A function is a block of reusable code that performs a specific task.
Instead of writing the same code again and again, you write it once and call it whenever needed.

üëâ Simple definition:

A function is a named piece of code that runs when it is called.

### 2. Why Do We Use Functions

- ‚úÖ Avoid code repetition

- ‚úÖ Make code clean & readable

- ‚úÖ Easy to debug and maintain

- ‚úÖ Divide a big problem into small parts

üîπ Basic Syntax of a Function

Example

```python
def greet():
    print("Hello, welcome to Python!")

greet()   # function call
```

### 4. Function with Return Value
return sends a result back to the caller.

```python
def square(n):
    return n * n
    # print(n * n)

result = square(4)
print(result)

# print(result)
# print(result)
```

### 5. Keyword Arguments
Arguments passed using parameter names.

```python
def info(name, age):
    print(name, age)

info(age=20, name="Ali")
```

### 6. *args (Variable Arguments)
Used when number of arguments is unknown.

```python
def total(*numbers):
    print(sum(numbers))

total(1, 2, 3, 4)
```

### 7. **kwargs (Keyword Variable Arguments)
Accepts key-value pairs.

```python
def student(**data):
    print(data)

student(name="Sajad", age=21, course="Python")
```

### 8. Lambda (Anonymous) Function
Small one-line function.

```python
square = lambda x: x * x
print(square(5))
```

### 02. Comparison Table: Modulus % vs Division 

| Feature     | `%` Modulus        | `/` Division    |
| ----------- | ------------------ | --------------- |
| Purpose     | Finds remainder    | Finds quotient  |
| Output type | Integer            | Float           |
| Example     | `7 % 3 = 1`        | `7 / 3 = 2.33`  |
| Used for    | Digits, even/odd   | Calculations    |
| Common use  | Last digit, checks | Math operations |


### 03. Comparison Table: Digit Extraction Techniques

| Task                | Formula       | Example     | Output |
| ------------------- | ------------- | ----------- | ------ |
| Last digit          | `n % 10`      | `123 % 10`  | `3`    |
| Remove last digit   | `n // 10`     | `123 // 10` | `12`   |
| First digit (logic) | Loop / string | `"123"[0]`  | `1`    |
| Sum of digits       | `%` + `//`    | `123 ‚Üí 6`   | `6`    |
