### What is a Dictionary

Definition:
A dictionary is a collection of data stored in key‚Äìvalue pairs.
Each key is used to access its value.

üëâ In simple words:
‚ÄúA dictionary stores data like a real dictionary‚Äîword (key) and meaning (value).‚Äù

### Important Features of Dictionaries
* Written using curly braces { }
* Data is stored as key : value
* Keys must be unique
* Values can be changed (mutable)
* Fast data access using keys

### Creating a Dictionary

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

print(student)
```

### Accessing Dictionary Values

Definition:
Values are accessed using keys, not index numbers.

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

print(student["name"])
print(student["marks"])
```

### Adding New Items

Definition:\
New key‚Äìvalue pairs can be added anytime.

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

student["grade"] = "A"
print(student)
```

### Changing Values

Definition:\
Values of existing keys can be updated.

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

student["age"] = 21
print(student)
```

### Removing Items

‚ùå pop()

Definition: Removes an item using its key.

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

student.pop("marks")
print(student)
```

‚ùå del

Definition: Deletes a key‚Äìvalue pair.

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

del student["age"]
print(student)
```

‚ùå clear()

Definition: Removes all items.

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

student.clear()
print(student)
```

### Common Dictionary Methods

üîπ keys()

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

print(student.keys())
```

üîπ items()

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

print(student.items())
```

üîπ get()

Definition: Gets value safely (no error if key not found).

```python
student = {
    "name": "Sajad ullah ",
    "age": 19,
    "marks": 85
}

print(student.get("name"))
```
