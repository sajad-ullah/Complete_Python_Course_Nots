### 0. What are Conditional Statements

Definition:
Conditional statements are used to make decisions in a program.
They allow the program to execute different code blocks based on whether a condition is True or False.

üëâ In simple words:
‚ÄúIf a condition is true, do this; otherwise, do something else.‚Äù

### if Statement
üìò Definition

The if statement executes a block of code only when the condition is True.

üß™ Code Example

```python
age = 18

if age >= 18:
    print("You are eligible to vote")
```

### if-else Statement
üìò Definition

The else statement runs when the if condition is False.

üß™ Code Example

```python
age = 16

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

### elif Statement
üìò Definition

elif (else if) is used to check multiple conditions when the first condition is False.

üß™ Code Example

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

### Conditional Statement Flow

- **if condition**
  - ‚úÖ True ‚Üí run `if` block
  - ‚ùå False ‚Üí
    - **elif condition**
      - ‚úÖ True ‚Üí run `elif` block
      - ‚ùå False ‚Üí
        - **else** ‚Üí run `else` block


### Teaching Summary
| Statement | Purpose                                |
| --------- | -------------------------------------- |
| if        | Executes when condition is True        |
| elif      | Checks another condition               |
| else      | Executes when all conditions are False |
